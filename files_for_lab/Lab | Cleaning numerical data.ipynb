### 7. Define a function that differentiates between continuous and discrete variables. Hint: A number of unique values might be useful. Store continuous data into a continuous variable and do the same for discrete and categorical.

```python
data.dtypes
```

```python
data.dtypes['customer']
```

```python
def check(data):
    
    # if the column's data.dtype == 'float' --> continuous
    
    c = data.columns
    continuous = []
    discrete = []
    categorical = []
    
    for i in c:
        if data.dtypes[i] == float:
            continuous.append(i)  #str
    
    #elif the column's data.dtype == 'int' --> discrete
    
        elif data.dtypes[i] == int:
            discrete.append(i)
    
        
    #else: categorical
    
        else:
            
            categorical.append(i)
    
    return continuous, discrete, categorical
    
        
```

```python
check(data)
```
